```python
import numpy as np

# Generate a list of numbers from 0 to 99
x = np.arange(0,100)

# Multiply values of x with 2 to get y
y = x*2

# Calculate square of values in for variable z
z = x**2

# Print x, y and z
print (x, y, z)
```

```python
# import matplotlib.pyplot and set inline plotting 
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise 1

Perform the following steps in the cell below:

* Create a new figure object `fig` using `.figure()` function.
* Use `add_axes()` to add an axis `ax` to the canvas at absolute location [0,0,1,1].
* Plot (x,y) on that axes and set the labels and title.

## Exercise 2

Perform following actions:

* Create a subplots figure with 3 rows and 4 columns
* Plot the lines $y=x$, $y=2x$, $y=3x$, $y=4x$,...$y=10x$, $y=11x$, $y=12x$ in the respective subplots over the domain [0,10].

```python
# line space is between 0 and 10
x = np.linspace(0, 10)
# figures and axes = 3 rows, 4 columns, figure size of 15 x 10
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15,10))
#the title
plt.title('Graphs')
#okay, so for 1, 12
for n in range(1,13):
#the n-1 is to accomodate for counting from 0 and placement. // for getting an integer
    row = (n-1)//4
#same work here with the columns, using a modulo
    col = n%4-1
#choosing axes
    ax = axes[row][col]
# the actual problem we're plotting
    y = [x*n for x in x]
#plot x and y
    ax.plot(x,y)
#here are titles for the different plots 
    ax.set_title('{}x'.format(n))
```

## Exercise 3

Repeat the above exercise, but standardize the axes of all of your subplots so that you can more easily compare the slopes of the lines.


```python
# figures and axes = 3 rows, 4 columns, figure size of 15 x 10
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15,10))
#the title
plt.title('Graphs')
#okay, so for 1, 12
for n in range(1,13):
#the n-1 is to accomodate for counting from 0 and placement. // for getting an integer
    row = (n-1)//4
#same work here with the columns, using a modulo
    col = n%4-1
#choosing axes
    ax = axes[row][col]
# the actual problem we're plotting
    y = [xn*n for xn in x]
#plot x and y
    ax.plot(x,y)
#limits for axes
    ax.set_xlim(0, 10), ax.set_ylim(0,120)
#here are titles for the different plots 
    ax.set_title('{}x'.format(n))
```

## Exercise 4

Perform following steps in the cell below:

* Create a figure of size 8x6 inches
* Add two axes using relative subplots to the figure by dividing it in 1 row and 2 columns
* Plot (x,y) and (x,z) on the ax1 and ax2 respectively. 
* Set the line width of first axes to 3, line style as dotted and color it red.
* Set the line width of second axes to 5, line style as dash-dot (-.) and color it blue.
* Give the plots some labels and titles

```python
# figures and axes = 1 rows, 2 columns, figure size of 8x6
# add two axes using relative subplots to the figure by dividing it in 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,6))

ax1 = axes[0]
ax2 = axes[1]

ax.set_xlabel('variable - x')
ax.set_ylabel('variable - y')
ax.set_title ('Left Plot')

ax2.set_xlabel('variable - x')
ax2.set_ylabel('variable - z')
ax2.set_title ('Right Plot')

# Plot (x,y) and (x,z) on the ax1 and ax2 respectively. 
ax1.plot(x, y, label='x, y', color='red', linewidth=3, linestyle = ':')
ax2.plot(x, z, label='x, z', color='blue', linewidth=5, linestyle = '-.')
```

## Exercise 5

Above figure looks fine but a bit out of proportion. Let's resize this to make the plots look more appealing by ensuring that subplots are square in shape. Also change the line style of first plot (left) and change the type of 2nd plot (right) to a scatter plot with a `^` marker style.

```python
# figures and axes = 1 rows, 2 columns, figure size of 8x6
# add two axes using relative subplots to the figure by dividing it in 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1 = axes[0]
ax2 = axes[1]

ax.set_aspect(1.0)

ax.set_xlabel('variable - x')
ax.set_ylabel('variable - y')
ax.set_title ('Left Plot')

ax2.set_xlabel('variable - x')
ax2.set_ylabel('variable - z')
ax2.set_title ('Right Plot')


# Plot (x,y) and (x,z) on the ax1 and ax2 respectively. 
ax1.plot(x, y, label='x, y', color='red')
ax2.scatter(x, z, label='x, z', marker='^', color='blue')
```
